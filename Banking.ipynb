```python

bank = {
       'user' : ['Simran' , 'Rahul' , 'Ravi' ],
        'accno' : [1001 , 1002 , 1003],
        'password' : ['1234' , '5678' , '9876'],
        'bal' : [1000.45 , 5000.00 , 5432.65] }
from getpass import getpass
admin="root"
admin_pass = "root@123"
while True:
    print("*"*50)
    print("WELCOME")
    print("*"*50)
    print("\n\n")
    print("1.\tLogin\n2.\tSignup\n3.\tAdimn\n4.\tExit")
    ch=int(input("\nEnter your choice: "))
    if ch == 1:
        accno=int(input("Enter your account number : "))
        index=bank['accno'].index(accno)
        if accno in bank['accno']:
            paswrd=input("Enter your password : ")
            if paswrd in bank['password'][index]:
                while True:
                    print("\n1.\tCredit\n2.\tDebit\n3.\tCheck balance\n4.\tPassword change\n5.\tLogout")
                    c = int(input("Enter your choice : "))
                    if c == 1:
                        n = int(input("Enter the number of amount you want to credit : "))
                        bal = bank['bal'][index]-n
                        print("Your current balance is : ",bal)
                        bank['bal'][index]=bal
                    elif c == 2:
                        n = int(input("Enter the number of amount you want to debit : "))
                        bal = bank['bal'][index]+n
                        print("Your current balance is : ",bal)
                        bank['bal'][index]=bal
                    elif c == 3:
                        print("Your current balance is : ", bank['bal'][index])
                    elif c == 4:
                        passwrd = getpass("Enter your old password : ")
                        if passwrd == paswrd:
                            newpass = getpass("Enter your new password : ")
                            checkpass = getpass("Enter your new password again : ")
                            if newpass == checkpass:
                                print("Your password has been changed successfully!!")
                                bank['password'][index] = checkpass
                            else :
                                print("Your password has not been matched")
                        else:
                            print("Your old password has not been matched")
                    elif c == 5:
                        print("You have being successfully loging out ")
                        break
                    else:
                        print("Invalid choice!")
    elif ch == 2:
        username = input("\nEnter your username : ")
        passwrd = input("\nEnter your password : ")
        bal = input("\nEnter your balance : ")
        accno = bank ['accno'][-1] + 1
        print("\nYour account number is : ", accno)
        bank['user'].append(username)
        bank['accno'].append(accno)
        bank['password'].append(passwrd)
        bank['bal'].append(bal)
    elif ch == 3:
        username = input("\nEnter your name: ")
        if username == admin:
            password = getpass("\nPassword: ")
            if password == admin_pass:
                for key,value in bank.items():
                    print()
                    print("{} = {}".format(key,value))
                    print()
            else:
                print("\nCheck your password! \n Try again!")
        else:
            print("\nCheck your username and Try again!")
    elif ch == 4:
        choice = input("\nDo you want to continue the exit (y/n) : ").strip().lower()
        if choice == 'y' or choice == 'yes':
            break
        else:
            continue
    else:
        print("Invalid Choice! ")
```
